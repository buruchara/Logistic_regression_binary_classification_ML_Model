```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
#Read the data
data = pd.read_csv('User_Data.csv')
x = data.iloc[:, [2,3]].values #x = data.iloc[:, 2:4].values
y = data.iloc[:, 4].values
x
```

```python
#Split/Divide the data into 2 parts, training and Testing Data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
x_train
```

```python
#Preprocessing and scale the features
from sklearn.preprocessing import StandardScaler
std_sc = StandardScaler()
x_train = std_sc.fit_transform(x_train)
x_test = std_sc.fit_transform(x_test)
x_train
```

```python
#Fitting Logistic Regression Classifier to training data
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```

```python
#Prediction of result
y_pred = classifier.predict(x_test)
df = pd.DataFrame({'Actual Values': y_test, 'Predicted Values': y_pred})
df
```

```python
#Making The confusion Matrix
import seaborn as sns

cm = confusion_matrix(y_pred, y_test)
cm
print(cm)
sns.heatmap(cm)
```
